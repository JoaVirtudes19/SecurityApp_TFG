```python
import cv2
from PIL import Image
import numpy
video_input = 0  #Path en caso de un video
capture = cv2.VideoCapture(video_input)
history = [0]*40 ###Historial de fotogramas
while (capture.isOpened()):
    ret, frame = capture.read()
    img = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    img_conv = Image.fromarray(img)
    pred_dict  = model_type.end2end_detect(img_conv, valid_tfms, model, class_map=class_map, detection_threshold=0.5)
    if 1 in pred_dict['detection']['label_ids']:
        history = history[1:]+[1]
    else:
        history = history[1:]+[0]

    if 1 in pred_dict['detection']['label_ids'] and sum(history) == 10:
        print(pred_dict['detection']['bboxes'])
        fecha = datetime.datetime.now()
        mensaje = "ARMA DETECTADA\n"+str(fecha.day)+"/"+str(fecha.month)+"/"+str(fecha.year)+"\nCantidad: "+str(len(pred_dict['detection']['label_ids']))
        for cid in cids:
            bot.send_message(cid,mensaje)
            bot.send_photo(cid, pred_dict['img'])
    nimg = np.array(pred_dict['img'])
    ocvim = cv2.cvtColor(nimg, cv2.COLOR_RGB2BGR)
    cv2.imshow('CCTV',ocvim)
    if (cv2.waitKey(1) == ord('q')):
        break

capture.release()
cv2.destroyAllWindows()
```
